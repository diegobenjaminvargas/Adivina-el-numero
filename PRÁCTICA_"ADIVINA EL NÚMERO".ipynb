```python
import random
import getpass
import os
import openpyxl
import pandas as pd
import datetime

def generar_numero():
    return random.randint(1, 1000)

def adivinar_numero(numero_a_adivinar, intentos):
    for intento in range(1, intentos + 1):
        intento_usuario = int(input(f"Intento {intento}/{intentos}. Adivina el número: "))
        if intento_usuario == numero_a_adivinar:
            print("¡Felicidades! Has adivinado el número.")
            return True
        elif intento_usuario < numero_a_adivinar:
            print("El número buscado es mayor.")
        else:
            print("El número buscado es menor.")
    print("Lo siento, has superado el número de intentos.")
    return False

def guardar_resultados(fecha, nombre_jugador, resultado, dificultad, modo_juego):
    if not os.path.exists("Estadísticas_Adivina_el_Numero.xlsx"):
        wb = openpyxl.Workbook()
        wb.active.append(["Fecha y hora", "Jugador", "Resultado", "Dificultad", "Modo de juego"])
        wb.save("Estadísticas_Adivina_el_Numero.xlsx")

    wb = openpyxl.load_workbook("Estadísticas_Adivina_el_Numero.xlsx")
    hoja = wb.active
    
    print("Fecha y hora:", fecha)
    
    try:
        fecha_formateada = fecha.strftime("%Y-%m-%d %H:%M")
    except AttributeError:
        print("Error: fecha no es un objeto datetime válido.")
        return
    
    hoja.append([fecha, nombre_jugador, resultado, dificultad, modo_juego])
    wb.save("Estadísticas_Adivina_el_Numero.xlsx")

    
def partida_solitario(dificultad_inicial):
    fecha = datetime.datetime.now()
    modo_juego = "Partida Solitario"
    dificultad_seleccionada = dificultad_inicial

    while True:
        if dificultad_seleccionada in ["Fácil", "Medio", "Difícil"]:
            break
        else:
            dificultad_seleccionada = input("Elige la dificultad (Fácil, Medio, Difícil): ")

    if dificultad_seleccionada == "Fácil":
        intentos = 20
    elif dificultad_seleccionada == "Medio":
        intentos = 12
    elif dificultad_seleccionada == "Difícil":
        intentos = 5
    else:
        print("Dificultad no válida.")
        return

    numero_a_adivinar = generar_numero()
    nombre_jugador = input("Introduce tu nombre: ")
    if adivinar_numero(numero_a_adivinar, intentos):
        resultado = "Victoria"
    else:
        resultado = "Derrota"
    guardar_resultados(fecha, nombre_jugador, resultado, dificultad_seleccionada, modo_juego)


    
def partida_2_jugadores():
 
    fecha = datetime.datetime.now()
    modo_juego = "Partida 2 jugadores"
    
    while True:
        dificultad = input("Elige la dificultad (Fácil, Medio, Difícil): ")
        if dificultad in ["Fácil", "Medio", "Difícil"]:
            break
        else:
            print("Dificultad no válida.")

    if dificultad == "Fácil":
        intentos = 20
    elif dificultad == "Medio":
        intentos = 12
    elif dificultad == "Difícil":
        intentos = 5

    print("Jugador 1: Ingresa el número a adivinar (entre 1 y 1000): ")
    while True:
        numero_a_adivinar = getpass.getpass(prompt="Número a adivinar: ")
        if numero_a_adivinar.isdigit() and 1 <= int(numero_a_adivinar) <= 1000:
            break
        else:
            print("Número no válido. Introduce un número entre 1 y 1000.")

    numero_a_adivinar = int(numero_a_adivinar)

    nombre_jugador = input("Introduce tu nombre (Jugador 2): ")
    if adivinar_numero(numero_a_adivinar, intentos):
        resultado = "Victoria"
    else:
        resultado = "Derrota"
    guardar_resultados(fecha, nombre_jugador, resultado, dificultad, modo_juego)


def estadistica():
    if os.path.exists("Estadísticas_Adivina_el_Numero.xlsx"):
        df = pd.read_excel("Estadísticas_Adivina_el_Numero.xlsx", index_col=0)  
        df.index = pd.to_datetime(df.index) 
        print("Estadísticas:")
        print(df)
        print("")
    else:
        print("No hay datos de estadísticas disponibles.")


def main():
    while True:
        print("¡ADIVINA EL NÚMERO!")
        print("1. Partida modo solitario")
        print("2. Partida 2 Jugadores")
        print("3. Estadística")
        print("4. Salir")
        opcion = input("Elige una opción: ")

        if opcion == "1":
            dificultad = input("Elige la dificultad (Fácil, Medio, Difícil): ")
            partida_solitario(dificultad)
        elif opcion == "2":
            partida_2_jugadores()
        elif opcion == "3":
            estadistica()
        elif opcion == "4":
            print("¡Hasta luego!")
            break
        else:
            print("Opción no válida.")

if __name__ == "__main__":
    main()
```
